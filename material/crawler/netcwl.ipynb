```python
from urllib.request import urlopen, Request, urlretrieve
from bs4 import BeautifulSoup
import redis

url = 'https://www.net-fashion.net/category/1781'
r = Request(url)
r.add_header("user-agent", "Mozilla/5.0")
response = urlopen(r)
html = BeautifulSoup(response)
# print(html)

# content2 = html.find_all('img', width="228")
content = html.find_all('img', class_="js-product-img")
content2 = html.find_all('a', target="_parent")



# print(content)


name = []
img = []
for i in content:

    name.append(i['alt'])

    img.append(i['src'])


src = []
for n in content2[2:]:

    src.append(n['href'])
    
srcc = []
for num in range(1, int(len(src)/2+1)):
    srcc.append(src[num*2-1])

print((srcc))

r = redis.Redis(host='redis', port=6379, decode_responses=True)

for ii in range(len(content)):
    r.set('NET_T-shirt_F_' + name[ii] + '_' + str(ii), name[ii] + '___' + img[ii] + '___' + srcc[ii])
    r.expire('NET_T-shirt_F_' + name[ii] + '_' + str(ii), 108000)
    
    
```
